### PREPARING FOR FEATURE ENGINEERING

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import warnings 
warnings.filterwarnings("ignore")
import pickle
import json
```

```python
df_train = pd.read_csv("Train_data.csv")
```

```python
pd.set_option("display.max_columns", None)
```

```python
df_train.head()
```

```python
for col in df_train.columns:
    if col!="class":
        print(col, len(df_train[col].value_counts()))
```

```python
df_train.head()
```

```python
drop_columns = ["num_outbound_cmds","is_host_login","land"]

"""
# Drop columns which have 0 variance.

df_train.drop(drop_columns, axis=1, inplace=True)
"""
```

```python
categorical_columns = ["protocol_type", "service", "flag"]
```

```python
numerical_columns = list(df_train.drop(categorical_columns+["class"],axis=1).columns)
```

Encoding categorical variables.

```python
label_encoder_dict = {}
for col in categorical_columns:
    label_encoder_dict.update({f"le_{col}":LabelEncoder()})
    label_encoder_dict[f"le_{col}"].fit(df_train[[col]])
```

Standard Scaling

```python
standard_scaler_dict = {}
for col in numerical_columns:
    standard_scaler_dict.update({f"ss_{col}":StandardScaler()})
    standard_scaler_dict[f"ss_{col}"].fit(df_train[[col]])
```

Save, load and test pickle.

```python
# Save the models as pickle
with open("label_encoder_dict.pickle", "wb") as le_pickle:
    pickle.dump(label_encoder_dict, le_pickle, protocol=pickle.HIGHEST_PROTOCOL)

with open("standard_scaler_dict.pickle", "wb") as ss_pickle:
    pickle.dump(standard_scaler_dict, ss_pickle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Load pickles

with open("label_encoder_dict.pickle", "rb") as le_pickle:
    label_encoder_dict = pickle.load(le_pickle)
    
with open("standard_scaler_dict.pickle", "rb") as ss_pickle:
    standard_scaler_dict = pickle.load(ss_pickle)
```

```python
label_encoder_dict["le_protocol_type"].classes_
```

```python
standard_scaler_dict["ss_diff_srv_rate"].feature_names_in_
```

Create configuration.json

```python
# Creating configuration json file.
configuration = {"categorical_columns":categorical_columns,"numerical_columns":numerical_columns,
                 "drop_columns":drop_columns,"label_encoder_path":"label_encoder_dict.pickle",
                 "standart_scaler_path":"standard_scaler_dict.pickle"}

json_str = json.dumps(configuration)
json_file = open("configuration.json", "w")
json_file.write(json_str)
json_file.close()
```

```python
# Read json file 
with open('configuration.json', 'r') as open_file:
  
    # Reading from json file
    configuration = json.load(open_file)
```
